#### 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
完善原始折线图1(辅助显示层)

```python
import random
# 1.准备数据 
x=range(60) # 60分钟  range 左闭右开
y_temperature=[random.uniform(15,18) for i in x] # 温度
# print(y_temperature)


# 2.创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3.绘制折线图
plt.plot(x,y_temperature)
# 修改x，y刻度
x_label=["11点{}分".format(i) for i in x]
# print(x_label)
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,30,2)) # 设置步长控制刻度

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("某城市11点到12点温度变化")
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)

# 4.显示画像
plt.show()
```

#### 完善折线图 图像层
 两个城市的温度变化

```python
x=range(60)
y_dalian=[random.uniform(5,20) for i in x]
y_chengdu=[random.uniform(15,25) for i in x]

plt.figure(figsize=(25,8),dpi=80)

plt.plot(x,y_dalian,color='r',linestyle='-.',label='大连')
# 颜色 r g b w c m y k 
# 线 -,--,-.,:,'',
plt.plot(x,y_chengdu,color='b',label='成都')

plt.legend() # 显示图例 需要在图像中设置label

x_label=['11点{}分'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,25,5))

# 描述信息
plt.title("大连和成都11点至12点天气变化")
plt.xlabel("时间")
plt.ylabel("温度")

plt.grid(alpha=0.9)

plt.show()
```

#### 多个坐标系显示 
面向对象的画图方法

```python
x=range(60)
y_dalian=[random.uniform(5,20) for i in x]
y_chengdu=[random.uniform(15,25) for i in x]

# 创建画布
# plt.figure(figsize=(25,8),dpi=80)
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(25, 8), dpi=80)

axes[0].plot(x,y_dalian,color='r',linestyle='-.',label='大连')
axes[1].plot(x,y_chengdu,color='b',label='成都')

axes[0].legend() # 显示图例 需要在图像中设置label
axes[1].legend()

# 多光标操作 command + 鼠标点击行

x_label=['11点{}分'.format(i) for i in x]
axes[0].set_xticks(x[::5],x_label[::5])
axes[0].set_yticks(range(0,25,5))
axes[1].set_xticks(x[::5],x_label[::5])
axes[1].set_yticks(range(0,25,5))

# 描述信息
axes[0].set_title("大连11点至12点天气变化")
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[1].set_title("成都11点至12点天气变化")
axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")

axes[0].grid(alpha=0.9)
axes[1].grid(alpha=0.9)

plt.show()
```

#### 常见图形种类及意义 
- 折线图 plot , 也可画各种数学图像
- 散点图 scatter 关系/规律 判断变量直接是否存在数量关联趋势 展示分布规律 
- 柱状图 bar 统计不同类别的数量大小
- 直方图 histogram 连续性的数据展示
- 饼图 pie π 占比

```python
# 绘制正弦函数图
import numpy as np

# 准备数据
x=np.linspace(-10,10,1000) #  linspace是左闭右闭 为了数据密集 生成 -10到10 的一千个数
y=np.sin(x)

# 创建画布
plt.figure(figsize=(20,8),dpi=80)
# 绘制函数图
plt.plot(x,y)
plt.grid(alpha=0.9)
# 显示
plt.show()
```

```python
# 绘制二次函数
x=np.linspace(-1,1,1000)
y=2*x*x

plt.figure()

plt.plot(x,y)
plt.grid()

plt.show()
```

```python
# 需求 对比每部电影的票房收入

# 1 准备数据
movie_names = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴', '降魔传','追捕','七十七天','密战','狂兽','其它']
tickets = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]

# 2 创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3 绘制图像
x_tickes=range(len(movie_names))
plt.bar(x_tickes,tickets,color=['b','r','g','y','c','m','y','k','c','g','b'])

plt.xticks(x_tickes,movie_names)# 名字对应

plt.title("电影票房收入对比")
plt.grid(alpha=0.5)

# 4 显示图像
plt.show()
```

#### 直方图
##### 直方图与柱状图的区别
- 直方图展示数据的分布，柱状图比较数据的大小。
- 直方图为定量数据，柱状图X轴为分类数据。
- 直方图柱子无间隔，柱状图柱子有间隔
- 直方图宽度可不一，柱状图柱子宽度须一致

#### 绘图时注意
- 注意组距，组距绘制会影响直方图呈现出来的数据分布，因此在绘制直方图的时候需要多次尝试改变组距
- 注意Y轴所代表的变量，Y轴上的变量可以是频次（数据出现了多少次）、频率（频数/总次数）、频率/组距，不同的变量会让直方图描述的数据分布意义不同
