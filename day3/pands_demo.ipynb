#### 设置新索引

```python
df=pd.DataFrame({'month':[1,4,7,10],
                 'year':[2022,2021,2020,2019],
                 'sale':[55,77,22,45]})
```

```python
df
```

```python
# 设置月份为新的索引
df.set_index('month',drop=False) #drop 设置是否扔掉month列
```

```python
df.set_index('month',drop=False).index
```

```python
# 设置多个索引
new_df=df.set_index(['month','year'])
```

```python
new_df
```

```python
new_df.index
```

```python
new_df.index.names 
```

```python
new_df.index.levels # 每个level的元组值
```
